**1.** –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –æ—Ü–µ–Ω–∫—É $\widehat{\theta}$ –º–µ—Ç–æ–¥–æ–º –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –ø–æ —è–≤–Ω–æ–π —Ñ–æ—Ä–º—É–ª–µ.

**2.** –í—ã–ø–∏—à–∏—Ç–µ —Ñ–æ—Ä–º—É–ª—ã –∏—Ç–µ—Ä–∞—Ü–∏–π –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ (GD) –∏ —Å—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ (SGD) –¥–ª—è –ø–æ–∏—Å–∫–∞ $\widehat{\theta}$. –ù–µ –∑–∞–±—É–¥—å—Ç–µ —É–∫–∞–∑–∞—Ç—å, –∫–∞–∫ –≤ SGD –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –±–∞—Ç—á (–Ω–∞–±–æ—Ä –æ–±—ä–µ–∫—Ç–æ–≤ –Ω–∞ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏).

**3.** –û–ø–∏—à–∏—Ç–µ, —á—Ç–æ –ø—Ä–æ–∏–∑–æ–π–¥–µ—Ç, –µ—Å–ª–∏ –¥–ª—è –Ω–µ–∫–æ—Ç–æ—Ä–æ–≥–æ $i$ –æ—à–∏–±–∫–∞ $\varepsilon_i$ –±—É–¥–µ—Ç –æ—á–µ–Ω—å –±–æ–ª—å—à–æ–π.


*–ó–∞–º–µ—á–∞–Ω–∏—è.*
1. –≠—Ç–æ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –∑–∞–¥–∞—á–∞, –Ω–µ –ø–µ—Ä–µ–ø—É—Ç–∞–π—Ç–µ üòÑ
2. –í –æ—Ç–≤–µ—Ç–µ –≤–æ –≤—Å–µ—Ö –ø—É–Ω–∫—Ç–∞—Ö –º–æ–≥—É—Ç –æ—Å—Ç–∞–≤–∞—Ç—å—Å—è —Å—É–º–º—ã $n$ —Å–ª–∞–≥–∞–µ–º—ã—Ö, –Ω–æ –Ω–µ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –æ–ø–µ—Ä–∞—Ü–∏–π —Å –≤–µ–∫—Ç–æ—Ä–∞–º–∏ –∏–ª–∏ –º–∞—Ç—Ä–∏—Ü–∞–º–∏.
3. **–ü—Ä–æ—á–∏—Ç–∞–π—Ç–µ —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è –∫ —Ñ–æ—Ä–º–∞—Ç—É —Ä–µ—à–µ–Ω–∏—è –≤ –ø—Ä–∞–≤–∏–ª–∞—Ö —Å–¥–∞—á–∏ –∑–∞–¥–∞–Ω–∏—è.**

**–†–µ—à–µ–Ω–∏–µ:**

### 1.  **–û—Ü–µ–Ω–∫–∞ $\hat{\theta}$ –º–µ—Ç–æ–¥–æ–º –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ (–ú–ù–ö).**

–ù–∞ –ª–µ–∫—Ü–∏–∏ –±—ã–ª–∞ –ø–æ–ª—É—á–µ–Ω–∞ –æ–±—â–∞—è —Ñ–æ—Ä–º—É–ª–∞ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –≤–µ–∫—Ç–æ—Ä–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $\theta$ –≤ –º–∞—Ç—Ä–∏—á–Ω–æ–º –≤–∏–¥–µ:

$\hat{\theta} = (X^T X)^{-1} X^T Y$

–í –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ, –º–æ–¥–µ–ª—å –æ–¥–Ω–æ–º–µ—Ä–Ω–æ–π –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ *–±–µ–∑ —Å–≤–æ–±–æ–¥–Ω–æ–≥–æ —á–ª–µ–Ω–∞* –∏–º–µ–µ—Ç –≤–∏–¥ $y(x) = \theta x$.  –ü–æ—ç—Ç–æ–º—É:

*   $X$ - —ç—Ç–æ –≤–µ–∫—Ç–æ—Ä-—Å—Ç–æ–ª–±–µ—Ü, —Å–æ—Å—Ç–æ—è—â–∏–π –∏–∑ –∑–Ω–∞—á–µ–Ω–∏–π $x_i$:  $X = \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix}$
*   $Y$ - —ç—Ç–æ –≤–µ–∫—Ç–æ—Ä-—Å—Ç–æ–ª–±–µ—Ü, —Å–æ—Å—Ç–æ—è—â–∏–π –∏–∑ –∑–Ω–∞—á–µ–Ω–∏–π $y_i$: $Y = \begin{bmatrix} y_1 \\ y_2 \\ \vdots \\ y_n \end{bmatrix}$
*   $\theta$ - —ç—Ç–æ —Å–∫–∞–ª—è—Ä.

–ü–æ–¥—Å—Ç–∞–≤–∏–º —ç—Ç–∏ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –æ–±—â—É—é —Ñ–æ—Ä–º—É–ª—É:

$X^T X = \begin{bmatrix} x_1 & x_2 & \dots & x_n \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix} = \sum_{i=1}^{n} x_i^2$

$(X^T X)^{-1} = \frac{1}{\sum_{i=1}^{n} x_i^2}$  (—Ç–∞–∫ –∫–∞–∫ —ç—Ç–æ —Å–∫–∞–ª—è—Ä)

$X^T Y = \begin{bmatrix} x_1 & x_2 & \dots & x_n \end{bmatrix} \begin{bmatrix} y_1 \\ y_2 \\ \vdots \\ y_n \end{bmatrix} = \sum_{i=1}^{n} x_i y_i$

–°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ:

$\hat{\theta} = (X^T X)^{-1} X^T Y = \frac{1}{\sum_{i=1}^{n} x_i^2} \cdot \sum_{i=1}^{n} x_i y_i = \frac{\sum_{i=1}^{n} x_i y_i}{\sum_{i=1}^{n} x_i^2}$

### 2. GD –∏ SGD

*   **GD:**

–û–±—â–∞—è —Ñ–æ—Ä–º—É–ª–∞ –∏—Ç–µ—Ä–∞—Ü–∏–∏ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞:

$\theta_{t+1} = \theta_t - \eta \sum_{i=1}^{n} x_i (x_i^T \theta_t - y_i)$

–ü–æ–¥—Å—Ç–∞–≤–∏–º –Ω–∞—à–∏ –∑–Ω–∞—á–µ–Ω–∏—è ($x_i$ - —Å–∫–∞–ª—è—Ä, $x_i^T \theta_t = \theta_t x_i$):

$\theta_{t+1} = \theta_t - \eta \sum_{i=1}^{n} x_i (\theta_t x_i - y_i) = \theta_t - \eta \left( \theta_t \sum_{i=1}^n x_i^2 - \sum_{i=1}^n x_i y_i \right)$


*   **SGD:**

–û–±—â–∞—è —Ñ–æ—Ä–º—É–ª–∞ –∏—Ç–µ—Ä–∞—Ü–∏–∏ —Å—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞:

$\theta_{t+1} = \theta_t - \eta \frac{n}{K} \sum_{j=1}^{K} x_{i_j} (x_{i_j}^T \theta_t - y_{i_j})$

–ü–æ–¥—Å—Ç–∞–≤–∏–º –Ω–∞—à–∏ –∑–Ω–∞—á–µ–Ω–∏—è ($x_{i_j}$ - —Å–∫–∞–ª—è—Ä, $x_{i_j}^T \theta_t = \theta_t x_{i_j}$):

$\theta_{t+1} = \theta_t - \eta \frac{n}{K} \sum_{j=1}^{K} x_{i_j} (\theta_t x_{i_j} - y_{i_j})$

### 3. –í–ª–∏—è–Ω–∏–µ –±–æ–ª—å—à–æ–π –æ—à–∏–±–∫–∏ $\epsilon_i$:

*   **–ú–ù–ö:** –°–∏–ª—å–Ω–æ —Å–º–µ—â–∞–µ—Ç –æ—Ü–µ–Ω–∫—É $\hat{\theta}$ –≤ —Å—Ç–æ—Ä–æ–Ω—É –≤—ã–±—Ä–æ—Å–∞ –∏–∑-–∑–∞ –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏–∏ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—à–∏–±–æ–∫. –ë–æ–ª—å—à–æ–µ $\epsilon_i$ –ø—Ä–∏–≤–æ–¥–∏—Ç –∫ –±–æ–ª—å—à–æ–º—É –ø–æ –º–æ–¥—É–ª—é –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—é $x_i y_i$ –≤ —á–∏—Å–ª–∏—Ç–µ–ª–µ —Ñ–æ—Ä–º—É–ª—ã –¥–ª—è $\hat{\theta}$.

*   **GD:** –í—ã–∑—ã–≤–∞–µ—Ç –±–æ–ª—å—à–æ–π "—Å–∫–∞—á–æ–∫" –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $\theta$ –Ω–∞ –∏—Ç–µ—Ä–∞—Ü–∏–∏, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–π –≤—ã–±—Ä–æ—Å—É.

*   **SGD:** –ú–µ–Ω–µ–µ —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª–µ–Ω –∫ –≤—ã–±—Ä–æ—Å–∞–º, —á–µ–º GD –∏ –ú–ù–ö. –í–ª–∏—è–Ω–∏–µ –≤—ã–±—Ä–æ—Å–∞ –ø—Ä–æ—è–≤–ª—è–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –µ—Å–ª–∏ –æ–Ω –ø–æ–ø–∞–¥–∞–µ—Ç –≤ —Ç–µ–∫—É—â–∏–π –±–∞—Ç—á. –ü—Ä–∏ –º–∞–ª–æ–º —Ä–∞–∑–º–µ—Ä–µ –±–∞—Ç—á–∞, –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è –≤—ã–±—Ä–æ—Å–∞ –º–µ–Ω—å—à–µ, –∏ SGD –±–æ–ª–µ–µ —É—Å—Ç–æ–π—á–∏–≤.

–ü–æ—Å–º–æ—Ç—Ä–∏—Ç–µ –ø–æ –≥—Ä–∞—Ñ–∏–∫–∞–º, –∫–∞–∫ *—Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç –æ—Å—Ç–∞–ª—å–Ω—ã—Ö* –∏ –ø–æ–π–º–∏—Ç–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä —ç—Ç–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏. –°—Ç–∞—Ä–∞–π—Ç–µ—Å—å —Å—Ç—Ä–æ–∏—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–≤–Ω—ã–µ –≥—Ä–∞—Ñ–∏–∫–∏, –∞ –∏–∑ —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ —Ä–µ—à–µ–Ω–∏—è —É–±–∏—Ä–∞—Ç—å —Ç–µ, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –ø—Ä–∏–≥–æ–¥–∏–ª–∏—Å—å.

–ö–∞–∫–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ, –∞ –∫–∞–∫–∏–µ &mdash; –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ? –£—á—Ç–∏—Ç–µ —ç—Ç–æ –ø—Ä–∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–∏ –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–≤–Ω—ã—Ö –≥—Ä–∞—Ñ–∏–∫–æ–≤.

*–£–∫–∞–∑–∞–Ω–∏–µ*. –°—Ä–µ–¥–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –µ—Å—Ç—å —Ç–∞–∫–∏–µ, –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ—Ç–æ—Ä—ã—Ö –∫–∞–∫-—Ç–æ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω—ã, –∏ –∏—Ö –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –Ω–µ–±–æ–ª—å—à–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ. –ß—Ç–æ–±—ã –ø–æ–Ω—è—Ç—å, —è–≤–ª—è—é—Ç—Å—è –ª–∏ –æ–Ω–∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–º–∏, –ø–æ–ø—Ä–æ–±—É–π—Ç–µ –ø–æ–Ω—è—Ç—å, –∫–∞–∫ —Å –∏—Ö –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –±—É–¥—É—Ç —Ä–∞–±–æ—Ç–∞—Ç—å –ª–∏–Ω–µ–π–Ω—ã–µ –º–æ–¥–µ–ª–∏. –ò–Ω–∞—á–µ –≥–æ–≤–æ—Ä—è, –¥–æ–ø—É—Å—Ç–∏–º–æ –ª–∏ —Å —Ç–æ—á–∫–∏ –∑—Ä–µ–Ω–∏—è –ø—Ä–∏—Ä–æ–¥—ã –¥–∞–Ω–Ω—ã—Ö –≤—ã–ø–æ–ª–Ω—è—Ç—å –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏ –º–µ–∂–¥—É *–ª—é–±—ã–º–∏* –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ —ç—Ç–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞, —Å–æ–≥–ª–∞—Å—É–µ—Ç—Å—è –ª–∏ —ç—Ç–æ —Å–æ —Å–º—ã—Å–ª–æ–º, –∫–æ—Ç–æ—Ä—ã–π –Ω–µ—Å–µ—Ç –≤ —Å–µ–±–µ –¥–∞–Ω–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫?

```python
feature_names = {
    'season': '–í—Ä–µ–º—è –≥–æ–¥–∞',
    'mnth': '–ú–µ—Å—è—Ü',
    'holiday': '–ü—Ä–∞–∑–¥–Ω–∏–∫ –ª–∏',
    'weekday': '–î–µ–Ω—å –Ω–µ–¥–µ–ª–∏',
    'workingday': '–†–∞–±–æ—á–∏–π –ª–∏ –¥–µ–Ω—å',
    'weathersit': '–ö–∞—á–µ—Å—Ç–≤–æ –ø–æ–≥–æ–¥—ã (–±–æ–ª—å—à–µ - —Ö—É–∂–µ)',
    'temp': '–¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞',
    'atemp': '–¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ –ø–æ –æ—â—É—â–µ–Ω–∏—è–º',
    'hum': '–í–ª–∞–∂–Ω–æ—Å—Ç—å',
    'windspeed': '–°–∫–æ—Ä–æ—Å—Ç—å –≤–µ—Ç—Ä–∞',
    'cnt': '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞—Ä–µ–Ω–¥'
}

season_names = {1: '–ó–∏–º–∞', 2: '–í–µ—Å–Ω–∞', 3: '–õ–µ—Ç–æ', 4: '–û—Å–µ–Ω—å'}
month_names = {1: '–Ø–Ω–≤', 2: '–§–µ–≤', 3: '–ú–∞—Ä', 4: '–ê–ø—Ä', 5: '–ú–∞–π', 6: '–ò—é–Ω',
               7: '–ò—é–ª', 8: '–ê–≤–≥', 9: '–°–µ–Ω', 10: '–û–∫—Ç', 11: '–ù–æ—è', 12: '–î–µ–∫'}
yes_no = {0: '–ù–µ—Ç', 1: '–î–∞'}

train_data_transformed = train_data.copy()
train_data_transformed['season'] = train_data_transformed['season'].map(season_names)
train_data_transformed['mnth'] = train_data_transformed['mnth'].map(month_names)
train_data_transformed['holiday'] = train_data_transformed['holiday'].map(yes_no)
train_data_transformed['workingday'] = train_data_transformed['workingday'].map(yes_no)
yes_no = {0: '–ù–µ—Ç', 1: '–î–∞'}

month_order = ['–Ø–Ω–≤', '–§–µ–≤', '–ú–∞—Ä', '–ê–ø—Ä', '–ú–∞–π', '–ò—é–Ω',
               '–ò—é–ª', '–ê–≤–≥', '–°–µ–Ω', '–û–∫—Ç', '–ù–æ—è', '–î–µ–∫']
season_order = ['–ó–∏–º–∞', '–í–µ—Å–Ω–∞', '–õ–µ—Ç–æ', '–û—Å–µ–Ω—å']

# –ì—Ä–∞—Ñ–∏–∫–∏ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ cnt –æ—Ç –∫–∞–∂–¥–æ–≥–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞
plt.figure(figsize=(15, 10))
plt.suptitle("–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∞—Ä–µ–Ω–¥ –æ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤", fontsize=16)

plot_features = [col for col in train_data.columns if col not in ['cnt', 'weekday']]

for i, col in enumerate(plot_features):
    plt.subplot(3, 3, i + 1)

    if train_data_transformed[col].dtype == 'int64' or train_data_transformed[col].dtype == 'float64':
        if len(train_data_transformed[col].unique()) <= 12:
            sns.boxplot(x=col, y='cnt', data=train_data_transformed)
        else:
            sns.regplot(x=col, y='cnt', data=train_data_transformed, scatter_kws={'alpha':0.3})
    else:
        if col == 'mnth':
             sns.boxplot(x=col, y='cnt', data=train_data_transformed, order=month_order)
        elif col == 'season':
            sns.boxplot(x=col, y='cnt', data=train_data_transformed, order=season_order)
        else:
            sns.boxplot(x=col, y='cnt', data=train_data_transformed)

    plt.xlabel(feature_names[col])
    plt.ylabel(feature_names['cnt'])

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

#### 2. –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö

–û–ø–∏—à–∏—Ç–µ –ø–æ —à–∞–≥–∞–º –ø—Ä–æ—Ü–µ–¥—É—Ä—É –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –≤–∫–ª—é—á–∞—Ç—å —Ä–∞–±–æ—Ç—É —Å –≤—ã–±—Ä–æ—Å–∞–º–∏, –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –≤–æ–∑–º–æ–∂–Ω–æ, —Å–æ–∑–¥–∞–Ω–∏–µ –∫–∞–∫–∏—Ö-–ª–∏–±–æ –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (–ø–æ—Å–º–æ—Ç—Ä–∏—Ç–µ <a href="https://thetahat.ru/courses/ad/main/3/ad_linreg_sklearn">–ø—Ä–∏–º–µ—Ä</a> —Å –∑–∞–Ω—è—Ç–∏—è). –¢–æ, –∫–∞–∫ –∏–º–µ–Ω–Ω–æ –≤—ã–ø–æ–ª–Ω—è—Ç—å –ø–æ–¥–≥–æ—Ç–æ–≤–∫—É –¥–∞–Ω–Ω—ã—Ö, —è–≤–ª—è–µ—Ç—Å—è —Ü–µ–ª—ã–º –∏—Å—Å–∫—É—Å—Ç–≤–æ–º. –ó–¥–µ—Å—å –Ω–µ—Ç "–ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ" –ø–æ–¥—Ö–æ–¥–∞, –∑–¥–µ—Å—å –µ—Å—Ç—å –º–µ—Ç—Ä–∏–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–º –º–Ω–æ–∂–µ—Å—Ç–≤–µ, –∫–æ—Ç–æ—Ä–∞—è –≤ –ø–æ—Å–ª–µ–¥—Å—Ç–≤–∏–∏ –æ—Ü–µ–Ω–∏—Ç –∫–∞—á–µ—Å—Ç–≤–æ –≤–∞—à–µ–π –º–æ–¥–µ–ª–∏.

*–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, –æ—Ç —Ç–æ–≥–æ, –Ω–∞—Å–∫–æ–ª—å–∫–æ –∫–∞—á–µ—Å—Ç–≤–µ–Ω–Ω–æ –≤—ã –ø—Ä–æ–≤–µ–¥–µ—Ç–µ –≤–∏–∑—É–∞–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –∏ –ø–æ–¥–≥–æ—Ç–æ–≤–∫—É –¥–∞–Ω–Ω—ã—Ö, –±—É–¥–µ—Ç –∑–∞–≤–∏—Å–µ—Ç—å –∫–∞—á–µ—Å—Ç–≤–æ –≤–∞—à–∏—Ö –º–æ–¥–µ–ª–µ–π –Ω–∞ —Å–ª–µ–¥—É—é—â–∏—Ö —à–∞–≥–∞—Ö.*

**–ü—Ä–æ—Ü–µ–¥—É—Ä–∞ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö:**

**a).** –ò—Ç–æ–≥–æ–≤–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ –ø—Ä–æ—Ü–µ–¥—É—Ä—ã –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö:

1.  **–û–±—Ä–∞–±–æ—Ç–∫–∞ –≤—ã–±—Ä–æ—Å–æ–≤ (–Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ):**

    *   –£–¥–∞–ª—è—é—Ç—Å—è –Ω–∞–±–ª—é–¥–µ–Ω–∏—è, –≥–¥–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞—Ä–µ–Ω–¥ (`cnt`) –ø—Ä–µ–≤—ã—à–∞–µ—Ç 6000 –∑–∏–º–æ–π (`season` = 1).
    *   –£–¥–∞–ª—è—é—Ç—Å—è –Ω–∞–±–ª—é–¥–µ–Ω–∏—è, –≥–¥–µ `cnt` –º–µ–Ω—å—à–µ 1900 –ª–µ—Ç–æ–º (`season` = 3).
    *   –£–¥–∞–ª—è—é—Ç—Å—è –Ω–∞–±–ª—é–¥–µ–Ω–∏—è, –≥–¥–µ `cnt` –º–µ–Ω—å—à–µ 700 –æ—Å–µ–Ω—å—é (`season` = 4).
    *   –í—ã–≤–æ–¥–∏—Ç—Å—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–¥–∞–ª–µ–Ω–Ω—ã—Ö —Å—Ç—Ä–æ–∫ –Ω–∞ –∫–∞–∂–¥–æ–º —ç—Ç–∞–ø–µ.

2.  **–ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (–Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ):**

    *   **–î–æ–±–∞–≤–ª–µ–Ω–∏–µ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞:** –î–æ–±–∞–≤–ª—è–µ—Ç—Å—è –∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫ `const`, –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω—ã–π –µ–¥–∏–Ω–∏—Ü–∞–º–∏.  –≠—Ç–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –¥–ª—è –º–æ–¥–µ–ª–µ–π –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å–æ —Å–≤–æ–±–æ–¥–Ω—ã–º —á–ª–µ–Ω–æ–º.

    *   **–°–æ–∑–¥–∞–Ω–∏–µ –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:**
        *   `season_sq`: –ö–≤–∞–¥—Ä–∞—Ç —Ä–∞–∑–Ω–æ—Å—Ç–∏ `(3 - season)`.  –≠—Ç–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –º–æ–∂–µ—Ç –ø–æ–º–æ—á—å –º–æ–¥–µ–ª–∏ —É–ª–æ–≤–∏—Ç—å –Ω–µ–ª–∏–Ω–µ–π–Ω—É—é –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É –≤—Ä–µ–º–µ–Ω–µ–º –≥–æ–¥–∞ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∞—Ä–µ–Ω–¥.  –í—ã—á–∏—Ç–∞–Ω–∏–µ –∏–∑ –∫–æ–Ω—Å—Ç–∞–Ω—Ç—ã 3 –≤—ã–±—Ä–∞–Ω–æ, —á—Ç–æ–±—ã –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø—Ä–∏—Ö–æ–¥–∏–ª–æ—Å—å –Ω–∞ –ª–µ—Ç–æ (—Å–µ–∑–æ–Ω 3).
        *   `mnth_abs_diff`: –ú–æ–¥—É–ª—å —Ä–∞–∑–Ω–æ—Å—Ç–∏ `abs(9 - mnth)`.  –≠—Ç–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Å–æ–∑–¥–∞–µ—Ç –ø—Ä–∏–∑–Ω–∞–∫, –∫–æ—Ç–æ—Ä—ã–π –æ—Ç—Ä–∞–∂–∞–µ—Ç —É–¥–∞–ª–µ–Ω–Ω–æ—Å—Ç—å –º–µ—Å—è—Ü–∞ –æ—Ç —Å–µ–Ω—Ç—è–±—Ä—è (–º–µ—Å—è—Ü 9). –°–µ–Ω—Ç—è–±—Ä—å –≤—ã–±—Ä–∞–Ω, –∫–∞–∫ –º–µ—Å—è—Ü, –±–ª–∏–∑–∫–∏–π –∫ –ø–∏–∫—É –∞—Ä–µ–Ω–¥—ã.
        *   `mnth_sq`: –ö–≤–∞–¥—Ä–∞—Ç —Ä–∞–∑–Ω–æ—Å—Ç–∏ `(9 - mnth)^2`. –ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ `mnth_abs_diff`, –Ω–æ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç —Ä–∞–∑–Ω–æ—Å—Ç–∏.
        *   `weathersit_sq`: –ö–≤–∞–¥—Ä–∞—Ç –∑–Ω–∞—á–µ–Ω–∏—è `weathersit`.  –≠—Ç–æ –º–æ–∂–µ—Ç –ø–æ–º–æ—á—å –º–æ–¥–µ–ª–∏ –ª—É—á—à–µ —É—á–µ—Å—Ç—å –≤–ª–∏—è–Ω–∏–µ –ø–ª–æ—Ö–æ–π –ø–æ–≥–æ–¥—ã (–±–æ–ª—å—à–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è `weathersit`).
        *   `temp_log`: –ù–∞—Ç—É—Ä–∞–ª—å–Ω—ã–π –ª–æ–≥–∞—Ä–∏—Ñ–º –æ—Ç `temp`.  –õ–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä–æ–≤–∞–Ω–∏–µ –º–æ–∂–µ—Ç –ø–æ–º–æ—á—å —Å–≥–ª–∞–¥–∏—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏ —Å–¥–µ–ª–∞—Ç—å –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –±–æ–ª–µ–µ –ª–∏–Ω–µ–π–Ω–æ–π.
        *   `atemp_log`: –ù–∞—Ç—É—Ä–∞–ª—å–Ω—ã–π –ª–æ–≥–∞—Ä–∏—Ñ–º –æ—Ç `atemp` (—Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ –ø–æ –æ—â—É—â–µ–Ω–∏—è–º). –ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ `temp_log`.

    *   **–£–¥–∞–ª–µ–Ω–∏–µ –∏—Å—Ö–æ–¥–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:** *–ü–æ—Å–ª–µ* —Å–æ–∑–¥–∞–Ω–∏—è –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –∏—Å—Ö–æ–¥–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ `season`, `mnth` –∏ `weekday` —É–¥–∞–ª—è—é—Ç—Å—è –∏–∑ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏, —á—Ç–æ–±—ã –∏–∑–±–µ–∂–∞—Ç—å –º—É–ª—å—Ç–∏–∫–æ–ª–ª–∏–Ω–µ–∞—Ä–Ω–æ—Å—Ç–∏ –∏ –∏–∑–±—ã—Ç–æ—á–Ω–æ—Å—Ç–∏ –¥–∞–Ω–Ω—ã—Ö.

–†–µ–∞–ª–∏–∑—É–π—Ç–µ –¥–∞–Ω–Ω—É—é –ø—Ä–æ—Ü–µ–¥—É—Ä—É.

```python
train_data_orig = train_data.copy()
```

```python
import numpy as np

train_data = train_data_orig.copy()

# 1. –û–±—Ä–∞–±–æ—Ç–∫–∞ –≤—ã–±—Ä–æ—Å–æ–≤ (–Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ):
print("1. –û–±—Ä–∞–±–æ—Ç–∫–∞ –≤—ã–±—Ä–æ—Å–æ–≤:")

# –ó–∏–º–∞: cnt > 6000
winter_outliers = (train_data['season'] == 1) & (train_data['cnt'] > 6000)
print(f"  - –£–¥–∞–ª–µ–Ω–æ –≤—ã–±—Ä–æ—Å–æ–≤ –∑–∏–º–æ–π (cnt > 6000): {winter_outliers.sum()}")
train_data = train_data[~winter_outliers]

# –õ–µ—Ç–æ: cnt < 1900
summer_outliers = (train_data['season'] == 3) & (train_data['cnt'] < 1900)
print(f"  - –£–¥–∞–ª–µ–Ω–æ –≤—ã–±—Ä–æ—Å–æ–≤ –ª–µ—Ç–æ–º (cnt < 1900): {summer_outliers.sum()}")
train_data = train_data[~summer_outliers]

# –û—Å–µ–Ω—å: cnt < 700
autumn_outliers = (train_data['season'] == 4) & (train_data['cnt'] < 700)
print(f"  - –£–¥–∞–ª–µ–Ω–æ –≤—ã–±—Ä–æ—Å–æ–≤ –æ—Å–µ–Ω—å—é (cnt < 700): {autumn_outliers.sum()}")
train_data = train_data[~autumn_outliers]


# 2. –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:
print("\n2. –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:")

# 2.1 –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞:
print("  - –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞")
train_data['const'] = 1

# 2.2 –°–æ–∑–¥–∞–Ω–∏–µ –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:
print("  - –°–æ–∑–¥–∞–Ω–∏–µ –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:")
train_data['season_sq'] = (3 - train_data['season'])**2
print("    - season_sq: (3 - season)^2")
train_data['mnth_abs_diff'] = abs(9 - train_data['mnth'])
print("    - mnth_abs_diff: abs(9 - mnth)")
train_data['mnth_sq'] = (9 - train_data['mnth'])**2
print("    - mnth_sq: (9 - mnth)^2")
train_data['weathersit_sq'] = train_data['weathersit']**2
print("    - weathersit_sq: weathersit^2")
train_data['temp_log'] = np.log(train_data['temp'])
print("    - temp_log: log(temp)")
train_data['atemp_log'] = np.log(train_data['atemp'])
print("    - atemp_log: log(atemp)")

# 2.3 –£–¥–∞–ª–µ–Ω–∏–µ –∏—Å—Ö–æ–¥–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ —Å–æ–∑–¥–∞–Ω–∏—è –Ω–æ–≤—ã—Ö:
print("  - –£–¥–∞–ª–µ–Ω–∏–µ –∏—Å—Ö–æ–¥–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ season, mnth, weekday")
train_data = train_data.drop(columns=['season', 'mnth', 'weekday'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model = LinearRegression()

X_train = train_data.drop('cnt', axis=1)
y_train = train_data['cnt']

model.fit(X_train, y_train)

coefficients = model.coef_
intercept = model.intercept_

print("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –º–æ–¥–µ–ª–∏:")
for feature, coef in zip(X_train.columns, coefficients):
    print(f"{feature}: {coef:.4f}")
print(f"–°–≤–æ–±–æ–¥–Ω—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç: {intercept:.4f}")
```

```python
# –í—ã—á–∏—Å–ª—è–µ–º –æ—Ü–µ–Ω–∫—É –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –ø–æ —è–≤–Ω–æ–π —Ñ–æ—Ä–º—É–ª–µ (X^T X)^-1 X^T y
X = X_train.values
y = y_train.values

theta_hat = np.linalg.inv(X.T @ X) @ X.T @ y

# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞–º–∏ –∏–∑ sklearn –≤ –≤–∏–¥–µ DataFrame
comparison_df = pd.DataFrame({
    '–ü—Ä–∏–∑–Ω–∞–∫': X_train.columns.tolist() + ['Intercept'],
    '–Ø–≤–Ω–∞—è —Ñ–æ—Ä–º—É–ª–∞': theta_hat.tolist() + [np.nan],
    'sklearn': coefficients.tolist() + [intercept]
})
comparison_df = comparison_df.set_index('–ü—Ä–∏–∑–Ω–∞–∫')

# –§–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –≤—ã–≤–æ–¥–∞: —É–±–∏—Ä–∞–µ–º —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω—É—é –∑–∞–ø–∏—Å—å –∏ —É–≤–µ–ª–∏—á–∏–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–Ω–∞–∫–æ–≤
pd.options.display.float_format = '{:.6f}'.format  # –ì–ª–æ–±–∞–ª—å–Ω–æ –¥–ª—è –≤—Å–µ–≥–æ DataFrame
print("\n–°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞–º–∏ –∏–∑ sklearn:")
print(comparison_df)
```

### –ó–∞–¥–∞—á–∞ 3.

*–ü–µ—Ä–µ–¥ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–µ–º –∑–∞–¥–∞—á–∏ –æ–∑–Ω–∞–∫–æ–º—Ç–µ—Å—å —Å <a href="https://thetahat.ru/courses/ad/main/3/ad_logreg_sklearn">–Ω–æ—É—Ç–±—É–∫–æ–º –ø–æ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏</a> —Å –∑–∞–Ω—è—Ç–∏—è.*

> –û–¥–Ω–æ –∏–∑ –∏–Ω—Ç–µ—Ä–µ—Å–Ω—ã—Ö —Å–≤–æ–π—Å—Ç–≤ –º–æ–¥–µ–ª–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ ‚Äî *–ø—Ä–∏ —Å–æ–±–ª—é–¥–µ–Ω–∏–∏ –µ—ë –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–π* –æ–Ω–∞ –¥–∞–µ—Ç –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∞—Ç—å **–Ω–µ—Å–º–µ—â–µ–Ω–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π** –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏ –æ–±—ä–µ–∫—Ç–∞ –∫ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–º—É –∫–ª–∞—Å—Å—É.  
> –î–ª—è –ø–æ—è—Å–Ω–µ–Ω–∏—è —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º –æ–±—ä–µ–∫—Ç $x$ –∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ $\widehat{p}(x)$ –¥–ª—è –∫–ª–∞—Å—Å–∞ 1. –ï—Å–ª–∏ –≤–∑—è—Ç—å –Ω–µ–±–æ–ª—å—à—É—é –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç—å –æ–±—ä–µ–∫—Ç–∞ $x$, —Ç–æ –¥–æ–ª—è –æ–±—ä–µ–∫—Ç–æ–≤ –∫–ª–∞—Å—Å–∞ 1 –≤ —ç—Ç–æ–π –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ –±—É–¥–µ—Ç –ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω–æ —Ä–∞–≤–Ω–∞ $\widehat{p}(x)$.  

–î–∞–ª–µ–µ –ø—Ä–æ–≤–µ—Ä–∏–º —ç—Ç–æ —Å–≤–æ–π—Å—Ç–≤–æ –Ω–∞ –∫–æ–Ω–∫—Ä–µ—Ç–Ω—ã—Ö –ø—Ä–∏–º–µ—Ä–∞—Ö.

–° –ø–æ–º–æ—â—å—é –∫–æ–¥–∞ –Ω–∏–∂–µ —Å–≥–µ–Ω–µ—Ä–∏—Ä—É–π—Ç–µ –¥–∞–Ω–Ω—ã–µ, —Å–æ—Å—Ç–æ—è—â–∏–µ –∏–∑ –æ–¥–Ω–æ–≥–æ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –∏ –±–∏–Ω–∞—Ä–Ω–æ–≥–æ —Ç–∞—Ä–≥–µ—Ç–∞.

```python
sample_size = 3000  # –†–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏

# –ü—Ä–∏–∑–Ω–∞–∫–∏
X = np.random.uniform(low=-4, high=4, size=(sample_size, 1))

# –¢–∞—Ä–≥–µ—Ç
y_mean_true = 1 / (1 + np.exp(1 - 2 * X.ravel()))
y = np.random.binomial(n=1, p=y_mean_true)

plt.figure(figsize=(10, 3))
plt.scatter(X, y_mean_true, marker=".", s=1, label="–†–µ–∞–ª—å–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏\n(–º—ã –∏—Ö –Ω–µ –∑–Ω–∞–µ–º)")
plt.scatter(X, y, marker="|", alpha=0.1, label="–î–∞–Ω–Ω—ã–µ")
plt.xlabel("–ü—Ä–∏–∑–Ω–∞–∫")
plt.ylabel("–ö–ª–∞—Å—Å –æ–±—ä–µ–∫—Ç–∞")
plt.legend();
```

–û–±—É—á–∏—Ç–µ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏—é, –∏—Å–ø–æ–ª—å–∑—É—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—é –∏–∑ `sklearn`, –ø—Ä–∏ —ç—Ç–æ–º —Å–≤–æ–±–æ–¥–Ω—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –¥–æ–ª–∂–µ–Ω –ø—Ä–∏—Å—É—Ç—Å—Ç–≤–æ–≤–∞—Ç—å –≤ –º–æ–¥–µ–ª–∏. –£–∫–∞–∂–∏—Ç–µ —Ç–∞–∫–∂–µ `penalty='none'`.

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(penalty=None, solver='lbfgs', max_iter=1000)
model.fit(X, y)
```

```python
print("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ø—Ä–∏ –ø—Ä–∏–∑–Ω–∞–∫–µ:", logreg.coef_[0][0])
print("–°–≤–æ–±–æ–¥–Ω—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç:", logreg.intercept_[0])
```

* –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –∫–ª–∞—Å—Å–æ–≤,
* –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –∫–ª–∞—Å—Å–∞ 1.

–í–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–π—Ç–µ —ç—Ç–∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è. –ù–∞ –≥—Ä–∞—Ñ–∏–∫ —Å—Ç–æ–∏—Ç –Ω–∞–Ω–µ—Å—Ç–∏ —Ç–∞–∫–∂–µ –æ–±—É—á–∞—é—â—É—é –≤—ã–±–æ—Ä–∫—É.

```python
x_grid = np.linspace(-4, 4, 10000).reshape((-1, 1))

y_pred_class = logreg.predict(x_grid)
y_pred_proba = logreg.predict_proba(x_grid)[:, 1]

plt.figure(figsize=(10, 6))

plt.scatter(X, y, marker='|', alpha=0.1, label='–û–±—É—á–∞—é—â–∞—è –≤—ã–±–æ—Ä–∫–∞', color='orange')
plt.scatter(x_grid, y_pred_class, marker='.', s=10, label='–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∫–ª–∞—Å—Å—ã', color='red')
plt.plot(x_grid, y_pred_proba, label='–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏', color='blue')

plt.xlabel('–ü—Ä–∏–∑–Ω–∞–∫')
plt.ylabel('–ö–ª–∞—Å')
plt.legend()
plt.title('–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏')
plt.grid(True)
plt.show()
```

–†–∞–∑–±–µ–π—Ç–µ –æ—Ç—Ä–µ–∑–æ–∫ $[-4, 4]$ –Ω–∞ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –±–∏–Ω—ã –¥–ª–∏–Ω—ã –¥–ª–∏–Ω—ã 0.2 –∏ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ –≤ –∫–∞–∂–¥–æ–º –±–∏–Ω–µ –¥–æ–ª—é –æ–±—ä–µ–∫—Ç–æ–≤ –∫–ª–∞—Å—Å–∞ 1. –ü–æ–ª—É—á–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–æ–±–∞–≤—å—Ç–µ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –∏ —Å—Ä–∞–≤–Ω–∏—Ç–µ —ç—Ç–∏ –≥—Ä–∞—Ñ–∏–∫–∏. –ü—Ä–æ–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä—É–π—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã.

*–ü–æ–¥—Å–∫–∞–∑–∫–∞:* –º–æ–∂–µ—Ç –ø–æ–º–æ—á—å `np.digitize` –∏ –º–µ—Ç–æ–¥ `groupby` –¥–ª—è —Ç–∞–±–ª–∏—Ü `pandas`. –†–µ–∫–æ–º–µ–Ω–¥—É–µ–º –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å <a href="https://thetahat.ru/courses/python">–æ–±—É—á–∞—é—â–∏–µ –Ω–æ—É—Ç–±—É–∫–∏</a> –ø–æ –±–∏–±–ª–∏–æ—Ç–µ–∫–∞–º.

```python
bin_width = 0.2
bins = np.arange(-4, 4 + bin_width, bin_width)
bin_indices = np.digitize(X.ravel(), bins)

df = pd.DataFrame({'X': X.ravel(), 'y': y, 'bin_index': bin_indices})

bin_means = df.groupby('bin_index')['y'].mean().reset_index()

bin_centers = bins[:-1] + bin_width / 2

bin_centers_matched = bin_centers[:len(bin_means)]

bin_means['bin_center'] = bin_centers_matched

plt.figure(figsize=(10, 6))

plt.plot(x_grid, y_pred_proba, label='–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏', color='green')

plt.scatter(bin_means['bin_center'], bin_means['y'], marker='o', color='blue', label='–î–æ–ª—è –∫–ª–∞—Å—Å–∞ 1 –≤ –±–∏–Ω–∞—Ö')

plt.xlabel('–ü—Ä–∏–∑–Ω–∞–∫')
plt.ylabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å / –î–æ–ª—è')
plt.legend()
plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –∏ –¥–æ–ª–µ–π –≤ –±–∏–Ω–∞—Ö')
plt.grid(True)
plt.show()
```

```python
# –ü—Ä–∏–∑–Ω–∞–∫–∏
X = np.random.uniform(low=-4, high=4, size=(sample_size, 1))

# –¢–∞—Ä–≥–µ—Ç
y_mean_true = 1 / (1 + np.exp(-100 * X.ravel()))
y = np.random.binomial(n=1, p=y_mean_true)

plt.figure(figsize=(10, 3))
plt.scatter(X, y_mean_true, marker=".", s=1, label="–†–µ–∞–ª—å–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏")
plt.scatter(X, y, marker="|", alpha=0.1, label="–î–∞–Ω–Ω—ã–µ")
plt.xlabel("–ü—Ä–∏–∑–Ω–∞–∫")
plt.ylabel("–ö–ª–∞—Å—Å –æ–±—ä–µ–∫—Ç–∞")
plt.legend();

logreg = LogisticRegression(penalty=None, fit_intercept=True, solver='lbfgs', max_iter=1000)
logreg.fit(X, y)

grid_size = 1000
x_grid = np.linspace(-4, 4, grid_size).reshape(-1, 1)
y_pred_proba = logreg.predict_proba(x_grid)[:, 1]
y_pred_class = logreg.predict(x_grid)

bin_width = 0.2
bins = np.arange(-4, 4 + bin_width, bin_width)
bin_indices = np.digitize(X.ravel(), bins)
df = pd.DataFrame({'X': X.ravel(), 'y': y, 'bin_index': bin_indices})
bin_means = df.groupby('bin_index')['y'].mean().reset_index()
bin_centers = bins[:-1] + bin_width / 2
bin_centers_matched = bin_centers[:len(bin_means)]
bin_means['bin_center'] = bin_centers_matched

plt.figure(figsize=(10, 6))
plt.scatter(X, y, marker='|', alpha=0.1, label='–û–±—É—á–∞—é—â–∞—è –≤—ã–±–æ—Ä–∫–∞')
plt.scatter(x_grid, y_pred_class, marker='.', s=10, label='–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∫–ª–∞—Å—Å—ã', color='red')
plt.plot(x_grid, y_pred_proba, label='–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏', color='blue')
plt.xlabel('–ü—Ä–∏–∑–Ω–∞–∫')
plt.ylabel('–ö–ª–∞—Å—Å / –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å')
plt.legend()
plt.title('–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏ (–Ω–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ)')
plt.grid(True)
plt.show()


plt.figure(figsize=(10, 6))
plt.plot(x_grid, y_pred_proba, label='–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏', color='green')
plt.scatter(bin_means['bin_center'], bin_means['y'], marker='o', color='blue', label='–î–æ–ª—è –∫–ª–∞—Å—Å–∞ 1 –≤ –±–∏–Ω–∞—Ö')
plt.xlabel('–ü—Ä–∏–∑–Ω–∞–∫')
plt.ylabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å / –î–æ–ª—è')
plt.legend()
plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –∏ –¥–æ–ª–µ–π –≤ –±–∏–Ω–∞—Ö (–Ω–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ)')
plt.grid(True)
plt.show()

print(f"–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ø—Ä–∏ –ø—Ä–∏–∑–Ω–∞–∫–µ: {logreg.coef_[0][0]}")
print(f"–°–≤–æ–±–æ–¥–Ω—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç: {logreg.intercept_[0]}")
```
