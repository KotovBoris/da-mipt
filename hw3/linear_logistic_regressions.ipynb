Рассмотрим модель одномерной линейной регрессии *без свободного параметра* $y(x) = \theta x$, где $x\in \mathbb{R}$ и $\theta \in \mathbb{R}$. Предполагаем, что данные получены по правилу

**Решение:**

### 1.  **Оценка $\hat{\theta}$ методом наименьших квадратов (МНК).**

На лекции была получена общая формула для оценки вектора параметров $\theta$ в матричном виде:

$\hat{\theta} = (X^T X)^{-1} X^T Y$

В нашем случае, модель одномерной линейной регрессии *без свободного члена* имеет вид $y(x) = \theta x$.  Поэтому:

*   $X$ - это вектор-столбец, состоящий из значений $x_i$:  $X = \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix}$
*   $Y$ - это вектор-столбец, состоящий из значений $y_i$: $Y = \begin{bmatrix} y_1 \\ y_2 \\ \vdots \\ y_n \end{bmatrix}$
*   $\theta$ - это скаляр.

Подставим эти значения в общую формулу:

$X^T X = \begin{bmatrix} x_1 & x_2 & \dots & x_n \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix} = \sum_{i=1}^{n} x_i^2$

$(X^T X)^{-1} = \frac{1}{\sum_{i=1}^{n} x_i^2}$  (так как это скаляр)

$X^T Y = \begin{bmatrix} x_1 & x_2 & \dots & x_n \end{bmatrix} \begin{bmatrix} y_1 \\ y_2 \\ \vdots \\ y_n \end{bmatrix} = \sum_{i=1}^{n} x_i y_i$

Следовательно:

$\hat{\theta} = (X^T X)^{-1} X^T Y = \frac{1}{\sum_{i=1}^{n} x_i^2} \cdot \sum_{i=1}^{n} x_i y_i = \frac{\sum_{i=1}^{n} x_i y_i}{\sum_{i=1}^{n} x_i^2}$

### 2. GD и SGD

*   **GD:**

Общая формула итерации градиентного спуска:

$\theta_{t+1} = \theta_t - \eta \sum_{i=1}^{n} x_i (x_i^T \theta_t - y_i)$

Подставим наши значения ($x_i$ - скаляр, $x_i^T \theta_t = \theta_t x_i$):

$\theta_{t+1} = \theta_t - \eta \sum_{i=1}^{n} x_i (\theta_t x_i - y_i) = \theta_t - \eta \left( \theta_t \sum_{i=1}^n x_i^2 - \sum_{i=1}^n x_i y_i \right)$


*   **SGD:**

Общая формула итерации стохастического градиентного спуска:

$\theta_{t+1} = \theta_t - \eta \frac{n}{K} \sum_{j=1}^{K} x_{i_j} (x_{i_j}^T \theta_t - y_{i_j})$

Подставим наши значения ($x_{i_j}$ - скаляр, $x_{i_j}^T \theta_t = \theta_t x_{i_j}$):

$\theta_{t+1} = \theta_t - \eta \frac{n}{K} \sum_{j=1}^{K} x_{i_j} (\theta_t x_{i_j} - y_{i_j})$

### 3. Влияние большой ошибки $\epsilon_i$:

*   **МНК:** Сильно смещает оценку $\hat{\theta}$ в сторону выброса из-за минимизации квадратов ошибок. Большое $\epsilon_i$ приводит к большому по модулю произведению $x_i y_i$ в числителе формулы для $\hat{\theta}$.

*   **GD:** Вызывает большой "скачок" параметра $\theta$ на итерации, соответствующей выбросу.

*   **SGD:** Менее чувствителен к выбросам, чем GD и МНК. Влияние выброса проявляется только если он попадает в текущий батч. При малом размере батча, вероятность попадания выброса меньше, и SGD более устойчив.

```python
feature_names = {
    'season': 'Время года',
    'mnth': 'Месяц',
    'holiday': 'Праздник ли',
    'weekday': 'День недели',
    'workingday': 'Рабочий ли день',
    'weathersit': 'Качество погоды (больше - хуже)',
    'temp': 'Температура',
    'atemp': 'Температура по ощущениям',
    'hum': 'Влажность',
    'windspeed': 'Скорость ветра',
    'cnt': 'Количество аренд'
}

season_names = {1: 'Зима', 2: 'Весна', 3: 'Лето', 4: 'Осень'}
month_names = {1: 'Янв', 2: 'Фев', 3: 'Мар', 4: 'Апр', 5: 'Май', 6: 'Июн',
               7: 'Июл', 8: 'Авг', 9: 'Сен', 10: 'Окт', 11: 'Ноя', 12: 'Дек'}
yes_no = {0: 'Нет', 1: 'Да'}

train_data_transformed = train_data.copy()
train_data_transformed['season'] = train_data_transformed['season'].map(season_names)
train_data_transformed['mnth'] = train_data_transformed['mnth'].map(month_names)
train_data_transformed['holiday'] = train_data_transformed['holiday'].map(yes_no)
train_data_transformed['workingday'] = train_data_transformed['workingday'].map(yes_no)
yes_no = {0: 'Нет', 1: 'Да'}

month_order = ['Янв', 'Фев', 'Мар', 'Апр', 'Май', 'Июн',
               'Июл', 'Авг', 'Сен', 'Окт', 'Ноя', 'Дек']
season_order = ['Зима', 'Весна', 'Лето', 'Осень']

# Графики зависимости cnt от каждого параметра
plt.figure(figsize=(15, 10))
plt.suptitle("Зависимость количества аренд от параметров", fontsize=16)

plot_features = [col for col in train_data.columns if col not in ['cnt', 'weekday']]

for i, col in enumerate(plot_features):
    plt.subplot(3, 3, i + 1)

    if train_data_transformed[col].dtype == 'int64' or train_data_transformed[col].dtype == 'float64':
        if len(train_data_transformed[col].unique()) <= 12:
            sns.boxplot(x=col, y='cnt', data=train_data_transformed)
        else:
            sns.regplot(x=col, y='cnt', data=train_data_transformed, scatter_kws={'alpha':0.3})
    else:
        if col == 'mnth':
             sns.boxplot(x=col, y='cnt', data=train_data_transformed, order=month_order)
        elif col == 'season':
            sns.boxplot(x=col, y='cnt', data=train_data_transformed, order=season_order)
        else:
            sns.boxplot(x=col, y='cnt', data=train_data_transformed)

    plt.xlabel(feature_names[col])
    plt.ylabel(feature_names['cnt'])

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

        *   `season_sq`: Квадрат разности `(3 - season)`.  Это преобразование может помочь модели уловить нелинейную зависимость между временем года и количеством аренд.  Вычитание из константы 3 выбрано, чтобы максимальное значение приходилось на лето (сезон 3).
        *   `mnth_abs_diff`: Модуль разности `abs(9 - mnth)`.  Это преобразование создает признак, который отражает удаленность месяца от сентября (месяц 9). Сентябрь выбран, как месяц, близкий к пику аренды.
        *   `mnth_sq`: Квадрат разности `(9 - mnth)^2`. Аналогично `mnth_abs_diff`, но использует квадрат разности.
        *   `weathersit_sq`: Квадрат значения `weathersit`.  Это может помочь модели лучше учесть влияние плохой погоды (большие значения `weathersit`).
        *   `temp_log`: Натуральный логарифм от `temp`.  Логарифмирование может помочь сгладить распределение и сделать зависимость более линейной.
        *   `atemp_log`: Натуральный логарифм от `atemp` (температура по ощущениям). Аналогично `temp_log`.

    *   **Удаление исходных признаков:** *После* создания новых признаков, исходные признаки `season`, `mnth` и `weekday` удаляются из обучающей выборки, чтобы избежать мультиколлинеарности и избыточности данных.

Реализуйте данную процедуру.

```python
train_data_orig = train_data.copy()
```

```python
import numpy as np

train_data = train_data_orig.copy()

# 1. Обработка выбросов (на обучающей выборке):
print("1. Обработка выбросов:")

# Зима: cnt > 6000
winter_outliers = (train_data['season'] == 1) & (train_data['cnt'] > 6000)
print(f"  - Удалено выбросов зимой (cnt > 6000): {winter_outliers.sum()}")
train_data = train_data[~winter_outliers]

# Лето: cnt < 1900
summer_outliers = (train_data['season'] == 3) & (train_data['cnt'] < 1900)
print(f"  - Удалено выбросов летом (cnt < 1900): {summer_outliers.sum()}")
train_data = train_data[~summer_outliers]

# Осень: cnt < 700
autumn_outliers = (train_data['season'] == 4) & (train_data['cnt'] < 700)
print(f"  - Удалено выбросов осенью (cnt < 700): {autumn_outliers.sum()}")
train_data = train_data[~autumn_outliers]


# 2. Преобразование признаков:
print("\n2. Преобразование признаков:")

# 2.1 Добавление константного признака:
print("  - Добавление константного признака")
train_data['const'] = 1

# 2.2 Создание новых признаков:
print("  - Создание новых признаков:")
train_data['season_sq'] = (3 - train_data['season'])**2
print("    - season_sq: (3 - season)^2")
train_data['mnth_abs_diff'] = abs(9 - train_data['mnth'])
print("    - mnth_abs_diff: abs(9 - mnth)")
train_data['mnth_sq'] = (9 - train_data['mnth'])**2
print("    - mnth_sq: (9 - mnth)^2")
train_data['weathersit_sq'] = train_data['weathersit']**2
print("    - weathersit_sq: weathersit^2")
train_data['temp_log'] = np.log(train_data['temp'])
print("    - temp_log: log(temp)")
train_data['atemp_log'] = np.log(train_data['atemp'])
print("    - atemp_log: log(atemp)")

# 2.3 Удаление исходных признаков после создания новых:
print("  - Удаление исходных признаков season, mnth, weekday")
train_data = train_data.drop(columns=['season', 'mnth', 'weekday'])
```
